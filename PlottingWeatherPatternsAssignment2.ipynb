```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
```

```python
#Splitting the Date into Years, Month-Date
df['Year'], df['MonthDate'] = zip(*df['Date'].apply(lambda x: (x[:4], x[5:])))
#Discarding leap year 29th Feb,only a standard of 365 days are considered
df = df[df['MonthDate'] != '02-29']
```

```python
import numpy as np
#Creating two Dataframes for Max & Min temperatures in 2015
temp_min_15 = df[(df['Element'] == 'TMIN') & (df['Year'] == '2015')].groupby('MonthDate').aggregate({'Data_Value': np.min})
temp_max_15 = df[(df['Element'] == 'TMAX') & (df['Year'] == '2015')].groupby('MonthDate').aggregate({'Data_Value':np.max})
```

```python
#Creating two Dataframes for Max & Min Values from 2005-14
temp_min = df[(df['Element'] == 'TMIN') & (df['Year'] != '2015')].groupby('MonthDate').aggregate({'Data_Value': np.min})
temp_max = df[(df['Element'] == 'TMAX') & (df['Year'] != '2015')].groupby('MonthDate').aggregate({'Data_Value':np.max})
```

```python
# When where the Max, Min value records broken?
broken_min = np.where(temp_min_15['Data_Value'] < temp_min['Data_Value'])[0]
broken_max = np.where(temp_max_15['Data_Value'] > temp_max['Data_Value'])[0]
```

```python
broken_min, broken_max
```

```python
import matplotlib as mpl
import matplotlib.dates as dates
import matplotlib.ticker as ticker
plt.figure()
fig, ax = plt.subplots( nrows=1, ncols=1, figsize = (10,5) ) 
plt.plot( temp_min.values, 'b', linewidth = 1, label = 'record low')
plt.plot(temp_max.values, 'r', linewidth = 1, label = 'record high')
plt.scatter(broken_min, temp_min_15.iloc[broken_min], s =10, c='g', label = 'broken low')
plt.scatter(broken_max, temp_max_15.iloc[broken_max], s =10, c='m', label = 'broken high')
plt.gca().axis([-5, 370, -150, 650])

# Graph Labels
plt.xlabel('Day of the Year 2015')
plt.ylabel('Temperature (Tenth of Degrees C)')
plt.title('Ann Arbor, Michigan Recorded Temperatures')
plt.suptitle("2015 temperature records vs past 10 years (2005-2014) records (°C)", fontsize=16)

#Graph Legend
plt.legend(loc = 1, frameon = False)

#Graph fill
plt.gca().fill_between(range(len(temp_min)), temp_min['Data_Value'], temp_max['Data_Value'], facecolor = 'pink', alpha = 0.5)

#Cleaning Y-Axis, limiting to achieve a relevant image in graph
ymin, ymax = -300, 450
ax.set_ylim(ymin, ymax)
ax.yaxis.set_major_locator(ticker.MultipleLocator(50)) # Putting major ticks at every 10 degrees Celsius
plt.show()
```
